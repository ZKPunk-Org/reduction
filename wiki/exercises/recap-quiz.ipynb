# Recap Quiz

:::{exercise}
:label: ex-recap-preference-order

Let's say you have a game $\mathsf{Game~}G$ that you want to prove is hard, and you have the option to upper bound $\Pr[\mathsf{Game~}G^{\mathcal{A}} = 1]$ by $\mathsf{Adv}^{\mathsf{AOMDL}}_{\mathcal{B}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{C}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{D}, \mathsf{GrGen}}$ in the ROM, or $\mathsf{Adv}^{\mathsf{DL}}_{\mathcal{E}, \mathsf{GrGen}}$.
What is your order of preference?

1. $\mathsf{Adv}^{\mathsf{DL}}_{\mathcal{E}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{AOMDL}}_{\mathcal{B}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{C}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{D}, \mathsf{GrGen}}$ in the ROM
2. $\mathsf{Adv}^{\mathsf{DL}}_{\mathcal{E}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{C}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{AOMDL}}_{\mathcal{B}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{D}, \mathsf{GrGen}}$ in the ROM
3. $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{D}, \mathsf{GrGen}}$ in the ROM, $\mathsf{Adv}^{\mathsf{DL}}_{\mathcal{E}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{AOMDL}}_{\mathcal{B}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{C}, \mathsf{GrGen}}$
4. $\mathsf{Adv}^{\mathsf{AOMDL}}_{\mathcal{B}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{DL}}_{\mathcal{E}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{C}, \mathsf{GrGen}}$, $\mathsf{Adv}^{\mathsf{OMDL}}_{\mathcal{D}, \mathsf{GrGen}}$ in the ROM
:::

:::{dropdown} **Solution**

The answer is (1).
:::

:::{exercise}
:label: ex-recap-game-hop

Consider the following game hop from $\mathsf{Game~Guess}$ to $\mathsf{Game~Guess}_2$:

**Game: Without the highlighted change, this is the guessing game. With the change, this is $\mathsf{Game~Guess}_2$.**

$$
\begin{array}{l}
\underline{\text{Game } \mathsf{Guess}_2:} \\
n \leftarrow 2^\lambda \\
x \xleftarrow{\$} \mathbb{Z}_n \\
\colorbox{yellow}{$\text{assert } x \neq 0$} \\
x' \leftarrow \mathcal{A}() \\
\text{return } x = x'
\end{array}
$$

Which of the following is correct?

1. $\left|\mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1]\right| \leq \Pr[x = 0]$
2. $\left|\mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1]\right| = \Pr[x = 0]$
3. $\left|\mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1]\right| \geq \Pr[x = 0]$
:::

:::{dropdown} **Solution**

The answer is (1).

By the Difference Lemma, when the games are identical except when event $E$ occurs (here $E$ is "$x = 0$"), the difference in success probabilities is bounded by $\Pr[E]$.
:::

:::{exercise}
:label: ex-recap-guess2-probability

Using the result from the previous exercise, what can we conclude about $\Pr[\mathsf{Guess}_2 = 1]$?

1. $\Pr[\mathsf{Guess}_2 = 1] \geq \mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - 2^{-\lambda}$
2. $\Pr[\mathsf{Guess}_2 = 1] \geq \mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} + 2^{-\lambda}$
3. $\Pr[\mathsf{Guess}_2 = 1] < \mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - 2^{-\lambda}$
:::

:::{dropdown} **Solution**

The answer is (1).

From the previous exercise, we have $\left|\mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1]\right| \leq 2^{-\lambda}$.

This means:

$$
-2^{-\lambda} \leq \mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1] \leq 2^{-\lambda}
$$

From the right inequality:

$$
\mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - \Pr[\mathsf{Guess}_2 = 1] \leq 2^{-\lambda}
$$

Rearranging:

$$
\Pr[\mathsf{Guess}_2 = 1] \geq \mathsf{Adv}^{\mathsf{Guess}}_{\mathcal{A}} - 2^{-\lambda}
$$
:::

:::{exercise}
:label: ex-recap-ro-single-query

Let $\mathcal{A}$ be an algorithm with access to a random oracle $\mathcal{O}^H: \{0,1\}^* \to \{0,1\}^\lambda$.
The algorithm queries the random oracle exactly once with some $x \in \{0,1\}^\lambda$.
What is $\Pr[\mathcal{O}^H(x) = x]$?

1. $0$
2. $2^{-\lambda}$
3. $1$
4. $2^{-2\lambda}$
:::

:::{dropdown} **Solution**

The answer is (2).

The random oracle returns a uniformly random value from $\{0,1\}^\lambda$ on the first query to any input.
Since there are $2^\lambda$ possible outputs and exactly one of them equals $x$, we have $\Pr[\mathcal{O}^H(x) = x] = 2^{-\lambda}$.
:::

:::{exercise}
:label: ex-recap-ro-two-queries

Let $p$ be a prime and let $\mathcal{A}$ be an algorithm with access to a random oracle $\mathcal{O}^H: \{0,1\}^* \to \mathbb{Z}_p$.
The adversary queries the random oracle exactly twice with distinct $x_1, x_2 \in \mathbb{Z}_p$ (in that order) and receives $y_1 = \mathcal{O}^H(x_1)$, $y_2 = \mathcal{O}^H(x_2)$.
What is $\Pr[y_2 = x_1 y_1 x_2^{-1}]$?

1. $0$
2. $1/p$
3. $1/(p-1)$
4. $1$
:::

:::{dropdown} **Solution**

The answer is (2).

Since $x_1 \neq x_2$, the queries are to different inputs, so $y_1$ and $y_2$ are independent uniform random values from $\mathbb{Z}_p$.
Given any fixed $x_1, x_2, y_1$, the value $x_1 y_1 x_2^{-1}$ is a fixed element of $\mathbb{Z}_p$.
Since $y_2$ is uniformly random and independent of $y_1$, we have $\Pr[y_2 = x_1 y_1 x_2^{-1}] = 1/p$.
:::

:::{exercise}
:label: ex-recap-taproot-collision

Let $\mathsf{gparam}$ be a group description.
Let $\mathcal{A}$ be an algorithm with access to a random oracle $\mathcal{O}^H: \{0,1\}^* \to \mathbb{Z}_p$ where $p$ is prime.
The adversary queries the random oracle exactly twice with $(R_0, m_0)$ and $(R_1, m_1)$ where $R_i \in \mathbb{G}$ and $m_i \in \{0,1\}^*$, receiving $y_0 = \mathcal{O}^H(R_0, m_0)$ and $y_1 = \mathcal{O}^H(R_1, m_1)$.
What is $\Pr[R_0 \cdot g^{y_0} \cdot R_1^{-1} = g^{y_1}]$?

1. $0$
2. $1/p$
3. $1/(p-1)$
4. $1$
:::

:::{dropdown} **Solution**

The answer is (2).

Since the queries $(R_0, m_0)$ and $(R_1, m_1)$ are distinct, $y_0$ and $y_1$ are independent uniform random values from $\mathbb{Z}_p$.

For any fixed $R_0, R_1 \in \mathbb{G}$ and $y_0 \in \mathbb{Z}_p$, the value $R_0 \cdot g^{y_0} \cdot R_1^{-1}$ is a fixed element of $\mathbb{G}$.
Since $y_1$ is uniformly random in $\mathbb{Z}_p$, $g^{y_1}$ is uniformly random in $\mathbb{G}$.
Therefore, $\Pr[R_0 \cdot g^{y_0} \cdot R_1^{-1} = g^{y_1}] = 1/p$.

**Note:** If this event occurred, we would have $R_0 \cdot g^{\mathcal{O}^H(R_0, m_0)} = R_1 \cdot g^{\mathcal{O}^H(R_1, m_1)}$, meaning

$$
\mathsf{TapCom.Commit}(R_0, m_0) = \mathsf{TapCom.Commit}(R_1, m_1)
$$

and the adversary would have won $\mathsf{Game~ComBind}^{\mathsf{TapCom}}$ in the random oracle model.
:::

:::{exercise}
:label: ex-recap-ro-table

The following demonstrates how games in the random oracle model can examine the table $T$ of previous queries.
Let $\mathsf{H}$ be a random oracle mapping to $\{0,1\}^\lambda$ and let $\mathcal{A}$ be an algorithm making at most $q_h$ queries to $\mathsf{H}$.
Let $\mathsf{Game~}G$ be the game defined below.
What is $\Pr[\mathsf{Game~}G = 1]$?

1. $0$
2. $1$
3. $\leq \frac{q_h}{2^\lambda}$
4. $\leq \frac{(q_h + 1) q_h}{2^\lambda}$

**Game $\mathsf{Game~}G$ for random oracle quiz question:**

$$
\begin{array}{ll}
\underline{\text{Game } G:} & \underline{\text{Oracle } \mathcal{O}^H(x):} \\
T \leftarrow \emptyset & \text{if } T[x] = \bot \text{ then} \\
(x, h) \leftarrow \mathcal{A}^{\mathcal{O}^H}() & \quad T[x] \xleftarrow{\$} \{0,1\}^\lambda \\
\text{if } \mathcal{O}^H(x) \neq h \text{ then} & \text{return } T[x] \\
\quad \text{return } \mathsf{false} & \\
\text{if } \exists x' \neq x : T[x'] = h \text{ then} & \\
\quad \text{return } \mathsf{true} & \\
\text{return } \mathsf{false} &
\end{array}
$$
:::

:::{dropdown} **Solution**

The answer is (3).

For the game to return $\mathsf{true}$, two conditions must be satisfied:

1. $\mathcal{O}^H(x) = h$ (the adversary's predicted hash value must be correct)
2. $\exists x' \neq x : T[x'] = h$ (there must be a collision: a different input $x'$ that was previously queried with the same hash $h$)

Since both conditions must hold:

$$
\Pr[\mathsf{Game~}G = 1] \leq \Pr[\exists x' \neq x : \mathcal{O}^H(x') = \mathcal{O}^H(x)]
$$

The adversary makes at most $q_h$ queries during its execution, and the game makes one additional query $\mathcal{O}^H(x)$.
So there are at most $q_h + 1$ total queries.
For each query, the probability of colliding with any of the at most $q_h$ entries already in the table is at most $\frac{q_h}{2^\lambda}$.

By the union bound over all $q_h + 1$ queries:

$$
\Pr[\exists x' \neq x : \mathcal{O}^H(x') = \mathcal{O}^H(x)] \leq (q_h + 1) \cdot \frac{q_h}{2^\lambda}
$$

Therefore, $\Pr[\mathsf{Game~}G = 1] \leq \frac{(q_h + 1)q_h}{2^\lambda}$.
:::

Which of the following represents a gap between security proofs and real-world security?

1. In practice, $\lambda$ is a fixed value (e.g., 128 or 256)
2. Concrete hash functions like SHA256 are not hash functions
3. Unfalsifiable assumptions cannot be validated experimentally
4. Security models may not capture all attack vectors
5. Random oracles do not exist
6. Security proofs don't model users writing keys on post-it notes
7. All of the above
:::

:::{dropdown} **Solution**

The answer is (7).
