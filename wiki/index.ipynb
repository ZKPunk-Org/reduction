[One-Way Function](https://en.wikipedia.org/wiki/One-way_function)
: A one-way function is a function that is easy to compute but hard to invert.

[Trapdoor Function](https://en.wikipedia.org/wiki/Trapdoor_function)
: A trapdoor function is a OWF ,yet only difficult to compute in the opposite direction without special information, called the "trapdoor".
