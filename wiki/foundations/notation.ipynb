# Notation

## Sets and Numbers

- $\mathbb{N}$ denotes the set of natural numbers $\{0, 1, 2, \ldots\}$.
- $\mathbb{Z}$ denotes the set of integers.
- $\mathbb{Z}_p$ denotes the set of integers modulo $p$, i.e., $\{0, 1, \ldots, p-1\}$.
- $\mathbb{Z}_p^*$ denotes the set of integers modulo $p$ excluding zero, i.e., $\{1, 2, \ldots, p-1\}$.
- $[n]$ denotes the set $\{1, 2, \ldots, n\}$ for a positive integer $n$.
- $[a, b]$ denotes the set $\{a, a+1, \ldots, b\}$ for integers $a \leq b$.
- $\{0, 1\}^n$ denotes the set of all binary strings of length $n$.
- $\{0, 1\}^*$ denotes the set of all binary strings of finite length.

## Probability and Sampling

- $x \leftarrow S$ denotes sampling $x$ uniformly at random from the finite set $S$.
- $x \leftarrow \mathcal{A}(y)$ denotes running algorithm $\mathcal{A}$ on input $y$ and assigning the output to $x$.
- $\Pr[E]$ denotes the probability of event $E$.
- $:=$ denotes a definition, i.e., the left side is defined to equal the right side.

## Cryptographic Notation

- $n$ denotes the security parameter.
- $1^n$ denotes the security parameter in unary representation.
- $\mathrm{negl}$ denotes a negligible function in the security parameter.
- PPT stands for probabilistic polynomial-time.

## Groups and Fields

- $(\mathbb{G}, g, p)$ denotes group parameters, where $\mathbb{G}$ denotes a cyclic group, $g$ denotes the generator and $p$ denotes the order of a group.
- $1_\mathbb{G}$ denotes the identity element of group $\mathbb{G}$.

## Logical Operators

- $\wedge$ denotes logical AND.
- $\vee$ denotes logical OR.
- $\neg$ denotes logical NOT.
- $\oplus$ denotes bitwise XOR.

## Other Notation

- $\|$ denotes concatenation (e.g., $x \| y$ is the concatenation of $x$ and $y$).
- $|S|$ denotes the cardinality (size) of set $S$.
- $\bot$ denotes an error or undefined value.
- $\emptyset$ denotes the empty set.
- $\log_g(h)$ denotes the discrete logarithm of $h$ with respect to base $g$.
- $O(\cdot)$ denotes asymptotic upper bound (big-O notation).
